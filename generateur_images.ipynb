```python
from PIL import Image, ImageDraw, ImageFilter
import random
```

```python

def generate_random_image(width, height, num_shapes=10):
    # Create a blank image with a white background
    image = Image.new("RGB", (width, height), "white")
    draw = ImageDraw.Draw(image)

    # Fill background with three random, similar colors
    background_colors = [
        (random.randint(0, 100), random.randint(0, 100), random.randint(0, 100)),
        (random.randint(100, 200), random.randint(100, 200), random.randint(100, 200)),
        (random.randint(200, 255), random.randint(200, 255), random.randint(200, 255)),
    ]
    draw.rectangle([0, 0, width, height], fill=random.choice(background_colors))

    # Define possible shapes (squares, circles, triangles)
    shapes = ["square", "circle", "triangle"]

    for _ in range(num_shapes):
        shape = random.choice(shapes)
        size = random.randint(20, 100)  # Adjust the size range as needed
        x = random.randint(0, width - size)
        y = random.randint(0, height - size)
        color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

        if shape == "square":
            draw.rectangle([x, y, x + size, y + size], fill=color)
        elif shape == "circle":
            draw.ellipse([x, y, x + size, y + size], fill=color)
        elif shape == "triangle":
            draw.polygon([(x, y), (x + size, y), (x + size // 2, y + size)], fill=color)

    return image

```

```python
# Example usage:
random_image = generate_random_image(500, 500, num_shapes=15)
random_image.show()
```
